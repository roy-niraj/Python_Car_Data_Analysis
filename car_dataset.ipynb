### 15) What is the minimum, maximum, and mean invoice price for sedans?


```python
car['Invoice'] = car['Invoice'].replace('[\\$,]', '', regex=True).astype(float)
car[car['Type'] == 'Sedan']['Invoice'].agg(['min', 'max', 'mean'])
```

### 16) What is the average weight of cars by origin (Asia, Europe, USA)?

```python

car['Weight'] = pd.to_numeric(car['Weight'],errors = 'coerce')
avg_Weight_by_origin = car.groupby('Origin')['Weight'].mean()
print(avg_Weight_by_origin)
```
